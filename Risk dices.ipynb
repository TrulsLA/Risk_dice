```python
# Risk Blitz Dice

print ('Welcome to Risk Blitz Dice!\n')

# importing libraries: Random, NumPy

import random
import numpy as np


# text library
attacker = 'Attacker:'
defender = 'Defender:'
dice = 'Dice '
attacker_won = 'Attacker won!'
defender_won = 'Defender won!'
error = '---ERROR---'


# attacker need to input no of attacking units, variable att_no number as integer
# defender need to input no of defending units, variable def_no number as integer

att_no = int(input('Attacker: How many attacking units? '))
def_no = int(input('Defender: How many defending units? '))
print()

# Test without input
#att_no = int(30)b
#def_no = int(20)

att_no_end = att_no
def_no_end = def_no

# printing summary

#print ('Attacker has ' + str(att_no) + ' units \nDefender has ' + str(def_no) + ' units.\n')


# generating two lists with random int 1-6, one for attacker att_dices and one for defender def_dices
# range defined by if statement from att_no and def_no

att_dices = []
def_dices = []



#range_att =
#range_def =

#if att_dices =<


for i in range(3):
    n = random.randint(1,6)
    att_dices.append(n)

for i in range(2):
    n = random.randint(1,6)
    def_dices.append(n)


# Sorting list in decending order

att_dices.sort(reverse=True)
def_dices.sort(reverse=True)


# Printing dice output sorted

print(attacker, end = ' ')
print (str(len(att_dices)) + ' dices ', end = '')
print(att_dices)
print(defender, end = ' ')
print(str(len(def_dices)) + ' dices ', end = '')
print(def_dices)


# Function for round x incl print result

def dice_round(dice_no):
    global att_no_end, def_no_end
    print('\n' + dice + str(dice_no + 1) + ': ' + str(att_dices[dice_no]) + ' vs ' + str(def_dices[dice_no]), end=' - ')
    if att_dices[dice_no] > def_dices[dice_no]:
        print (attacker_won, end = '')
        def_no_end -= 1
    elif att_dices[dice_no] <= def_dices[dice_no]:
        print (defender_won, end = '')
        att_no_end -= 1
    else:
        print(error, end = '')


# Execute functions, need to be adjusted for no of dices, while loop?

if att_dices == 1 or def_dices == 1:
    dice_round(0)
else:
    dice_round(0)
    dice_round(1)
    

# Test conclusion, use IF statements

print ('\n\nAttacker lost ' + str(att_no - att_no_end) + ' units, and now has '  + str(att_no_end) + ' units left.'
       + '\nDefender lost ' + str(def_no - def_no_end) + ' units, and now has '  + str(def_no_end) + ' units left.\n')
```
